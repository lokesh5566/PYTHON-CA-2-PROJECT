```python
import pandas as pd
df = pd.read_csv(r"C:\Users\methu\Downloads\Global College Statistics Dataset.csv")
print("Original shape:", df.shape)
negative_gender = df[(df['Male'] < 0) | (df['Female'] < 0)]
print("Rows with negative gender counts:\n", negative_gender)
df['Gender_Sum'] = df['Male'] + df['Female']
gender_mismatch = df[df['Gender_Sum'] != df['Total Students']]
print("Rows with gender mismatch:\n", gender_mismatch)
df_cleaned = df[(df['Male'] >= 0) & (df['Female'] >= 0)]
df_cleaned = df_cleaned[df_cleaned['Gender_Sum'] == df_cleaned['Total Students']]
df_cleaned.drop(columns=['Gender_Sum'], inplace=True)
print("Cleaned shape:", df_cleaned.shape)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\methu\Downloads\Global College Statistics Dataset.csv")
df['Placement Rate'] = pd.to_numeric(df['Placement Rate'], errors='coerce')
branch_avg_df = df.groupby('Branch', as_index=False)['Placement Rate'].mean()
branch_avg_df = branch_avg_df.sort_values(by='Placement Rate', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=branch_avg_df,x='Placement Rate',y='Branch',hue='Branch',palette='crest',dodge=False,legend=False)
plt.title('Average Placement Rate by Academic Branch')
plt.xlabel('Average Placement Rate (%)')
plt.ylabel('Academic Branch')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\methu\Downloads\Global College Statistics Dataset.csv") 
selected_cols = ['CGPA', 'Placement Rate', 'Annual Family Income','Research Papers Published', 'Faculty Count']
df[selected_cols] = df[selected_cols].apply(pd.to_numeric, errors='coerce')
df_clean = df[selected_cols].dropna()
sns.pairplot(df_clean, corner=True, diag_kind='kde', plot_kws={'alpha': 0.6, 'color': 'skyblue'})
plt.suptitle("Pair Plot of Academic Metrics", y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```
